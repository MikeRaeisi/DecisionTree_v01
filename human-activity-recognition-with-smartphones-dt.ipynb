# Break down the dataset into X and y

```python
X_train = df_train.copy()
y_train_cat = X_train.pop('Activity')
print('df_train.shape = ', df_train.shape)
print('X_train.shape = ', X_train.shape)
print('y_train_cat.shape = ', y_train_cat.shape)

X_test = df_test.copy()
y_test_cat = X_test.pop('Activity')
print('df_test.shape = ', df_test.shape)
print('X_test.shape = ', X_test.shape)
print('y_test_cat.shape = ', y_test_cat.shape)
```

```python
assert(X_train.columns == X_test.columns).all()
assert(y_train_cat.name == y_test_cat.name)
```

# Convert categorical to numerical

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

y_train = label_encoder.fit_transform(y_train_cat)
y_test = label_encoder.transform(y_test_cat)

y_train = pd.Series(y_train, index = y_train_cat.index, name = y_train_cat.name)
y_test = pd.Series(y_test, index = y_test_cat.index, name = y_test_cat.name)

# pd.get_dummies(y_train_cat)
```

# Column's relations

```python
train_cor = X_train.corrwith(y_train)
```

```python
from sklearn.feature_selection import mutual_info_classif

train_mi = mutual_info_classif(X_train, y_train, discrete_features='auto', n_neighbors=3, copy=True, random_state=None)
train_mi = pd.Series(train_mi, index = X_train.columns, name = 'Activity_MI')
```

```python
train_mi[train_mi==train_mi.max()]
```

# Model 1 - Decision Tree Classifier with some limitations

```python
from sklearn.tree import DecisionTreeClassifier

dtc_model = DecisionTreeClassifier(max_depth=150, min_samples_split = 5, min_samples_leaf = 3)

dtc_model.fit(X_train, y_train)

y_dtc_predict = dtc_model.predict(X_test)
y_dtc_predict = pd.Series(y_dtc_predict, index = X_test.index, name = 'pred_Activity')
```

```python
error_avg_dtc = (y_dtc_predict != y_test).sum()/len(y_test)
print(error_avg_dtc)
```

# Model 2 - Raw Decision Tree Classifier

```python
# from sklearn.tree import DecisionTreeClassifier

error_avg_dtc2 = 0
itr = 50

for i in range(itr):
    dtc_model2 = DecisionTreeClassifier()
    
    dtc_model2.fit(X_train, y_train)
    
    y_dtc2_predict = dtc_model2.predict(X_test)
    y_dtc2_predict = pd.Series(y_dtc2_predict, index = X_test.index, name = 'pred_Activity')
    
    error_avg_dtc2 += (y_dtc2_predict != y_test).sum()/len(y_test)

print(error_avg_dtc2/itr)
```

```python
# check overfitting
y_train_predict = dtc_model.predict(X_train)
y_train_predict = pd.Series(y_train_predict, index = X_train.index, name = 'pred_Activity')
error_avg_dtc2_train = (y_train_predict != y_train).sum()/len(y_train)
print(error_avg_dtc2_train)
```

# Model 3 - Decision Tree Classifier with some limitations

```python
# from sklearn.tree import DecisionTreeClassifier

error_avg_dtc3 = 0
itr = 50

for i in range(itr):
    dtc_model3 = DecisionTreeClassifier(max_depth=500, min_samples_split = 10, min_samples_leaf = 2)
    
    dtc_model3.fit(X_train, y_train)
    
    y_dtc3_predict = dtc_model3.predict(X_test)
    y_dtc3_predict = pd.Series(y_dtc3_predict, index = X_test.index, name = 'pred_Activity')
    
    error_avg_dtc3 += (y_dtc3_predict != y_test).sum()/len(y_test)

print(error_avg_dtc3/itr)
```

```python
# check for overfitting
y_train_predict3 = dtc_model3.predict(X_train)
y_train_predict3 = pd.Series(y_train_predict3, index = X_train.index, name = 'pred_Activity')
error_avg_dtc3_train = (y_train_predict3 != y_train).sum()/len(y_train)
print(error_avg_dtc3_train)
```

# Model 4 Extra Tree Classifier

```python
from sklearn.tree import ExtraTreeClassifier

error_avg_etc4 = 0
itr = 50
for i in range(itr):
    etc_model4 = ExtraTreeClassifier(criterion = 'gini', splitter = 'best', max_depth = 500, min_samples_split = 10, min_samples_leaf = 5, max_features = 'auto')
    
    etc_model4.fit(X_train, y_train)
    
    y_etc4_predict = etc_model4.predict(X_test)
    y_etc4_predict = pd.Series(y_etc4_predict, index = X_test.index, name = 'pred_Activity')
    
    error_avg_etc4 += (y_etc4_predict != y_test).sum()/len(y_test)
print(error_avg_etc4/itr)
```
